```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4])
m=0.5  # slope
c=4  # y intercept 
y=m*x+c
print(x,y)
plt.plot(x,y)
plt.scatter(x,y)
```

```python
area=np.array([700,750,800,900,1000])
price=np.array([20,30,33,45,75])
plt.scatter(area,price)
```

```python
m=1
c=2
price_pred=m*area+c
```

```python
# as we have taken random slope m and y intercept c value 
#so there is large distance between predicted and actual and 
#there will  big error or loss function 
# so we have to find actual values of  m and c 
# to get best fit line of predicted and actual = match of both 
# OLS= ordinary list square =to get bets fit line we use
plt.scatter(area,price)
plt.plot(area,price_pred)
```

```python
# OLS= ordinary list square =to get bets fit line we use
# for single linear regression equation= w0+W1x
# for multiple linear regression equation= w0+W1x1+w2x2+....Wnxn

# if y is actual value and y^ is predicted value then y-y^ is error called residue it can be +ve or -ve  so to find whole error we swuared this term y-y^ 

# total error= (y1-y1^)^2+ .......(yn-yn^)^2 
#MSE =mean square error= total error/n  it also called loss function of cost 
# and we have to find out the values of coefficient or weights that wiil give mininum value of MSE 
```

```python
print(area,price)
```

```python
mean_area=np.mean(area)
mean_price=np.mean(price)
print(mean_area,mean_price)
```

```python
W1=np.sum((area-mean_area)*(price-mean_price))/np.sum((area-mean_area)**2)
```

```python
W0=mean_price-W1*mean_area
```

```python
print(W1,W0)
```

```python
pred_price=W1*area+W0
```

```python
plt.scatter(area,price)
plt.plot(area,pred_price)
```
