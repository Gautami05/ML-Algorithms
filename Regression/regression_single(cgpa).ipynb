```python
# model building begins 
# now we have to split our data into training data and testing data
#random_state=0  it check random value for train 80% data
# test size = 0.2=> 20% ,  20=> 20 data points

from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)  # sequence will be  same train and test
```

```python
x_train.shape
x_train=x_train.values.reshape(-1,1)  # it is standard way to add dimensions arry,  converting into two dimension
```

```python
x_test.shape
x_train.shape
```

```python
# to fitting data into linear regression model we need atleast two dimension data 
# using linear regression 
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test.values.reshape(-1,1))
```
