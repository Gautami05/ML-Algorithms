```python
for i in mydata['make'].unique():     #for  loop company wise
    price=mydata[mydata['make']==i]['price']
    q1=np.quantile(price,0.25)
    q3=np.quantile(price,0.75)
    IQR=q3-q1
    #print(IQR)
    ub=q3+1.5*IQR
    index=price[price>ub].index
    print(index)
    if len(index)!=0:
        mydata.drop(index,inplace=True)
```

```python
for i in mydata['make'].unique():
    price=mydata[mydata['make']==i]['price']
    q1=np.quantile(price,0.25)
    q3=np.quantile(price,0.75)
    IQR=q3-q1
    #print(IQR)
    ub=q3+1.5*IQR
    lb=q1-1.5*IQR
    index=price[price>ub].index
    index2=price[price<lb].index
    print(index)
    if len(index)!=0:
        mydata.drop(index,inplace=True)
    if len(index2)!=0:
        mydata.drop(index2,inplace=True)
```

```python
for i in mydata['make'].unique():
    wheelbase=mydata[mydata['make']==i]['wheel-base']
    q1=np.quantile(wheelbase,0.25)
    q3=np.quantile(wheelbase,0.75)
    IQR=q3-q1
    #print(IQR)
    ub=q3+1.5*IQR
    lb=q1-1.5*IQR
    index=wheelbase[wheelbase>ub].index
    index2=wheelbase[wheelbase<lb].index
    print(index)
    if len(index)!=0:
        mydata.drop(index,inplace=True)
    if len(index2)!=0:
        mydata.drop(index2,inplace=True)
```

```python
# to handle sckeweness we can use srt or log transformation 
from scipy.stats import skew
print(skew(df_num['wheel-base']))
print(skew(df_num['horsepower']))
print(skew(df_num['price']))
print(skew(df_num['compression-ratio']))
```

```python
for i in df_num.columns:
    plt.figure()
    sns.displot(df_num[i],kind='kde')
    plt.show()
    print(skew(df_num[i]))
    
```

```python
# to find all the columns has skewness <1 and >-1
lst=[]
for i in df_num.columns:
    if skew(df_num[i])>1 or skew(df_num[i])<-1:
        lst.append(i)
print(lst)

```

```python
for i in lst:
    df_num[i]=np.log(df_num[i])
```
